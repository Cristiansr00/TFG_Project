```python
from ultralytics import YOLO
import cv2
import numpy as np
from skimage.filters.rank import entropy
from skimage.morphology import disk
import os
```

```python
# Load the model
model = YOLO("yolo11n.pt")

#Codigo para cargar la imagen original en 3 canales (RGB) o en este caso BGR
img = cv2.imread("2Gy-004.JPG")

#Mostramos la imagen que hemos cargado
cv2.imshow("Imagen Original", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Codigo para modificar la imagen a solo 1 canal (Blanco y negro)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#Mostramos la imagen en escala de grises
cv2.imshow("Imagen en escala de grises", img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Cargamos la imagen
img = cv2.imread("2Gy-004.JPG")
# Imagen en gris solo
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
b, g, r = cv2.split(img)
img2 = cv2.merge([b, g, r])

print(img2)

# cv2.imshow("B", b)
# cv2.imshow("G", g)
# cv2.imshow("R", r)
# cv2.imshow("img2", img2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# print("Shape original:", img.shape)      # (1024, 1280, 3)
# print("Shape gris:", img_gray.shape)     # (1024, 1280)
# cv2.imshow("Image", img)
# cv2.imshow("Grayscale Image", img_gray)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Binarización Otsu
_, binary_otsu = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

entropy_img = entropy(img_gray, disk(5))  # disk(5) = ventana de radio 5
# Normalizar a 0-255
entropy_norm = ((entropy_img - entropy_img.min()) / 
                (entropy_img.max() - entropy_img.min()) * 255).astype(np.uint8)

rgb = cv2.merge([b, entropy_norm, binary_otsu])        # Otsu en R


cv2.imshow("Original", img)
cv2.imshow("Binarizada Otsu", binary_otsu)
cv2.imshow("Entropia", entropy_norm)
cv2.imshow("Imagen rgb", rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()

# cv2.imwrite("2Gy-004_gray.jpg", img_gray)

# # ¿Son todos los canales iguales?
# same01 = np.array_equal(img[:,:,0], img[:,:,1])
# same02 = np.array_equal(img[:,:,0], img[:,:,2])
# print("canales 0==1:", same01, " 0==2:", same02)

# # Otra comprobación: número de píxeles donde hay diferencia entre canales
# diff01 = np.count_nonzero(img[:,:,0].astype(int) - img[:,:,1].astype(int))
# diff02 = np.count_nonzero(img[:,:,0].astype(int) - img[:,:,2].astype(int))
# print("píxeles distintos 0 vs 1:", diff01, " 0 vs 2:", diff02)
```
